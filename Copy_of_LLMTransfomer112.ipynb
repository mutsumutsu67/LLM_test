```python
import torch
from transformers import GPT2Config, GPT2LMHeadModel, GPT2Tokenizer, TextDataset, DataCollatorForLanguageModeling, Trainer, TrainingArguments

def create_and_train_model(model_name='gpt2', num_attention_layers=12, num_hidden_layers=2, train_data_path='train.txt', epochs=1):
    """
    新しいGPTモデルを作成し、事前学習を行う関数。
    :param model_name: 事前学習済みモデル名またはトークナイザのベースとなるモデル名。
    :param num_attention_layers: モデルのアテンション層の数。
    :param num_hidden_layers: モデルの隠れ層の数。
    :param train_data_path: 学習データファイルへのパス。
    :param epochs: 学習するエポック数。
    """

    # トークナイザの初期化
    tokenizer = GPT2Tokenizer.from_pretrained(model_name)

    # 日本語のトークナイズのために特別なトークンを追加
    tokenizer.add_special_tokens({
        "eos_token": "</s>",
        "bos_token": "<s>",
        "unk_token": "<unk>",
        "pad_token": "<pad>",
        "mask_token": "<mask>"
    })

    # モデル構成のカスタマイズ
    config = GPT2Config(
        vocab_size=len(tokenizer),
        n_layer=num_hidden_layers,
        n_head=num_attention_layers // num_hidden_layers,
        n_inner=4*num_attention_layers,
    )

    # モデルの初期化
    model = GPT2LMHeadModel(config)
    model.resize_token_embeddings(len(tokenizer))

    # データセットの準備
    train_dataset = TextDataset(
        tokenizer=tokenizer,
        file_path=train_data_path,
        block_size=128
    )

    data_collator = DataCollatorForLanguageModeling(
        tokenizer=tokenizer,
        mlm=False
    )

    # トレーニングの設定
    # トレーニングの設定
    training_args = TrainingArguments(
        output_dir="./gpt2_japanese",
        overwrite_output_dir=True,
        num_train_epochs=epochs,
        per_device_train_batch_size=4,
        save_steps=10_000,
        save_total_limit=2,
        logging_dir='./logs',  # ログを保存するディレクトリ
        logging_steps=10,  # 10ステップごとに進捗情報を出力
    )

    # トレーナーの初期化と学習の開始
    trainer = Trainer(
        model=model,
        args=training_args,
        data_collator=data_collator,
        train_dataset=train_dataset,
    )

    trainer.train()

# 学習データのパスとその他のパラメータを設定
create_and_train_model(
    model_name='gpt2',
    num_attention_layers=24,  # アテンション層の数を調整 12
    num_hidden_layers=8,      # 隠れ層の数を調整 2
    train_data_path='/content/drive/My Drive/working/Transfomer/kumono_ito2.txt',  # 学習データファイルへのパス
    epochs=40  # 学習エポック数 3
)


# train_data_path='path_to_your_training_data.txt',  # 学習データファイルへのパス


# epochs=3  # 学習エポック数
```

## 1. モデルのパラメータ数のカウント
モデルのパラメータ数をカウントすることで、その複雑さやおおよそのサイズを把握できます。PyTorch を使用している場合は、以下のようなコードで総パラメータ数と学習可能なパラメータ数を確認できます。

```python
import torch
from transformers import GPT2Model

# 例として GPT-2 モデルのロード（あなたのモデルに合わせてください）
model = GPT2Model.from_pretrained("gpt2")

# パラメータ数のカウント
total_params = sum(p.numel() for p in model.parameters())
trainable_params = sum(p.numel() for p in model.parameters() if p.requires_grad)

print(f"Total Parameters: {total_params}")
print(f"Trainable Parameters: {trainable_params}")
```

## 2. ディスク上のファイルサイズの確認
モデルをディスクに保存し、そのファイルサイズを直接確認することもできます。これには、まずモデルを保存し、次にそのファイルサイズを調べます。


```python
# モデルの保存
model_save_path = "./saved_model.bin"
torch.save(model.state_dict(), model_save_path)

# ファイルサイズの確認
import os
file_size = os.path.getsize(model_save_path)

print(f"Model File Size: {file_size / (1024*1024):.2f} MB")
```


このコードは、独自のGPTモデルを定義し、指定されたパラメータで初期化して、提供された訓練データで事前学習を行います。`create_and_train_model`関数はカスタマイズ可能で、特に`num_attention_layers`と`num_hidden_layers`を調整して、異なるモデルアーキテクチャを試すことができます。

### 重要な点
- 上記のコードでは、`train_data_path`に指定する訓練データは、事前に用意しておく必要があります。このパスは、モデルが訓練に使用するテキストファイルへのフルパスです。
- モデルサイズ（約10Mパラメータ）を目指すための正確な`num_attention_layers`と`num_hidden_layers`の値は、試行錯誤が必要です。上記の設定はあくまで一例です。モデルのサイズは、これらの層の数だけでなく、`n_inner`（FFN内の次元数）、`n_head`（アテンションヘッドの数）などの他のパラメータにも依存します。
- この例では、簡単化のために、訓練データを扱う部分が非常に基本的です。実際の使用では、データの前処理や後処理、さらには学習率の調整など、より洗練されたデータ処理やトレーニング戦略が必要になるかもしれません。
